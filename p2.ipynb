```python
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Step 2: Load the Iris dataset
iris = load_iris()
X = iris.data            # Features: shape (150, 4)
y = iris.target          # Labels: shape (150,)

# Step 3: Perform PCA to reduce from 4D to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 4: Print the first 5 rows of the transformed data
print("First 2 principal components of the first 5 samples:")
print(X_pca[:5])

# Step 5: Scatter plot
plt.figure(figsize=(8, 6))
for target, color, label in zip([0, 1, 2], ['r', 'g', 'b'], iris.target_names):
    plt.scatter(
        X_pca[y == target, 0],  # First PC
        X_pca[y == target, 1],  # Second PC
        color=color,
        label=label,
        edgecolor='k'
    )

plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA: Iris Dataset (2D Projection)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data        # Feature matrix (150 samples, 4 features)
y = iris.target      # Class labels (0, 1, 2)

# Apply PCA to reduce dimensionality from 4D to 2D
X_pca = PCA(n_components=2).fit_transform(X)

# Create a scatter plot of the 2D PCA projection
plt.scatter(
    X_pca[:, 0],               # First principal component (x-axis)
    X_pca[:, 1],               # Second principal component (y-axis)
    c=y,                       # Color by class label
    cmap='viridis',            # Color map
    edgecolor='k'              # Black border around points
)

# Add axis labels and title
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA: Iris Dataset (2D)")

# Show color bar for class labels
plt.colorbar()

# Display the plot
plt.show()
```
